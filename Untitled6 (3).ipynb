```python
# Full robust preprocessing script for your hybrid energy dataset
#WEEK1 MILESTONE


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import warnings
warnings.filterwarnings("ignore")

# ---------- 1. Load dataset (update path if different) ----------
csv_path = '/content/intermittent-renewables-production-france.csv'
df = pd.read_csv(csv_path)
print("Loaded dataset. Columns:", list(df.columns))
print("\nSample rows:")
display(df.head())
```

```python
#Data cleaning
# ---------- 2. Quick missing-values check ----------
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# ---------- 3. Create Datetime robustly ----------
def try_create_datetime(df):
    # strategy 1: if 'Datetime' already exists, try to parse it
    if 'Datetime' in df.columns:
        df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
        if df['Datetime'].notna().any():
            return df

    # strategy 2: try 'Date and Hour' column (already looks like "2020-07-22 20:00:00+02:00")
    if 'Date and Hour' in df.columns:
        df['Datetime'] = pd.to_datetime(df['Date and Hour'], errors='coerce')
        if df['Datetime'].notna().any():
            return df

    # strategy 3: combine 'Date' + 'StartHour' into a single string then parse
    if 'Date' in df.columns and 'StartHour' in df.columns:
        def normalize_hour(x):
            if pd.isna(x):
                return ''
            s = str(x).strip()

            if s.isdigit():
                return f"{int(s):02d}:00:00"

            try:
                if '.' in s and s.replace('.', '').isdigit():
                    return f"{int(float(s)):02d}:00:00"
            except:
                pass

            if ':' in s:
                parts = s.split(':')
                if len(parts) == 2:
                    return s + ":00"
                return s
            return s

        combined = df['Date'].astype(str).str.strip() + ' ' + df['StartHour'].apply(normalize_hour)
        df['Datetime'] = pd.to_datetime(combined, errors='coerce')
        if df['Datetime'].notna().any():
            return df

    for col in df.columns:
        try:
            test = pd.to_datetime(df[col], errors='coerce')
            if test.notna().sum() > 0.6 * len(df):  # >60% success
                df['Datetime'] = test
                return df
        except Exception:
            continue

    print("Warning: could not find/parse a datetime column reliably. Creating synthetic datetime index.")
    df['Datetime'] = pd.date_range(start='2000-01-01', periods=len(df), freq='H')
    return df

df = try_create_datetime(df)
print("\nAfter creation: Datetime null count =", int(df['Datetime'].isna().sum()))
```

```python
# ---------- 4. Force Datetime to pandas datetime dtype and remove tz info safely ----------
try:

    df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
except Exception:

    df['Datetime'] = pd.to_datetime(df['Datetime'].astype(str), utc=True, errors='coerce')


try:
    if df['Datetime'].dt.tz is not None:
        df['Datetime'] = df['Datetime'].dt.tz_convert('UTC').dt.tz_localize(None)
except Exception:

    try:
        df['Datetime'] = pd.to_datetime(df['Datetime'], utc=True, errors='coerce')
        df['Datetime'] = df['Datetime'].dt.tz_convert('UTC').dt.tz_localize(None)
    except Exception:
        if not pd.api.types.is_datetime64_any_dtype(df['Datetime']):
            df['Datetime'] = pd.date_range(start='2000-01-01', periods=len(df), freq='H')

print("Datetime dtype:", df['Datetime'].dtype)
display(df['Datetime'].head(8))
```

```python
# ---------- 5. Drop rows where 'Production' is missing ----------
if 'Production' in df.columns:
    missing_prod = int(df['Production'].isnull().sum())
    print(f"\nRows with missing 'Production': {missing_prod}")
    if missing_prod > 0:
        df = df.dropna(subset=['Production']).reset_index(drop=True)
        print("Dropped rows with missing Production.")
else:
    raise KeyError("Column 'Production' not found. Ensure CSV has 'Production' column.")
```

```python
# ---------- 6. Create time-based features (safe now that Datetime is datetime64) ----------
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
df['Month'] = df['Datetime'].dt.month
df['DayOfWeek'] = df['Datetime'].dt.dayofweek
```

```python
# ---------- 7. Pivot by Source (create Wind / Solar columns) ----------
if 'Source' in df.columns:

    pivot = df.groupby(['Datetime', 'Source'])['Production'].sum().unstack(fill_value=0).reset_index()
    pivot.columns.name = None
    print("\nPivoted data sample:")
    display(pivot.head(6))
else:
    pivot = df[['Datetime', 'Production']].rename(columns={'Production': 'Total_Production'})
    print("\nNo 'Source' column found; using Production as Total_Production.")
    display(pivot.head())
```

```python
# ---------- 8. Ensure Wind and Solar columns exist, create Total_Production ----------
if 'Wind' in pivot.columns or 'Solar' in pivot.columns:
    if 'Wind' not in pivot.columns:
        pivot['Wind'] = 0.0
    if 'Solar' not in pivot.columns:
        pivot['Solar'] = 0.0
    pivot['Total_Production'] = pivot['Wind'] + pivot['Solar']
else:

    numeric_cols = [c for c in pivot.columns if pd.api.types.is_numeric_dtype(pivot[c]) and c != 'Datetime']
    if 'Total_Production' not in pivot.columns:
        if numeric_cols:
            pivot['Total_Production'] = pivot[numeric_cols].sum(axis=1)
        else:

            pivot = pivot.merge(df[['Datetime','Production']].drop_duplicates('Datetime'), on='Datetime', how='left')
            pivot.rename(columns={'Production':'Total_Production'}, inplace=True)
            pivot['Total_Production'] = pivot['Total_Production'].fillna(0)
```

```python
# ---------- 9. Recreate time features on pivot and sort ----------
pivot = pivot.sort_values('Datetime').reset_index(drop=True)
pivot['Hour'] = pivot['Datetime'].dt.hour
pivot['Day'] = pivot['Datetime'].dt.day
pivot['Month'] = pivot['Datetime'].dt.month
pivot['DayOfWeek'] = pivot['Datetime'].dt.dayofweek

print("\nFinal processed data sample:")
display(pivot.head(6))
```

```python
# -------------------------
# Imports
# -------------------------
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib

from xgboost import XGBRegressor

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
# -------------------------
# 0. Load or create cleaned dataset
# -------------------------
CLEANED_PATH = '/content/cleaned_hybrid_energy.csv'
RAW_POSSIBLE = ['/content/intermittent-renewables-production-france.csv',
                '/content/wind-solar-electricity-production.csv']

if os.path.exists(CLEANED_PATH):
    print("Loading cleaned dataset:", CLEANED_PATH)
    df_clean = pd.read_csv(CLEANED_PATH, parse_dates=['Datetime'])
else:
    # try some raw paths
    raw_path = None
    for p in RAW_POSSIBLE:
        if os.path.exists(p):
            raw_path = p
            break
    if raw_path is None:
        print("Cleaned dataset not found. Please upload raw CSV file now (e.g., intermittent-renewables-production-france.csv)")
        from google.colab import files
        uploaded = files.upload()
        raw_path = list(uploaded.keys())[0]
    print("Loading raw dataset:", raw_path)
    df_raw = pd.read_csv(raw_path)
    # Minimal robust parsing similar to earlier preprocessing
    if 'Date and Hour' in df_raw.columns:
        df_raw['Datetime'] = pd.to_datetime(df_raw['Date and Hour'], errors='coerce')
    elif 'Date' in df_raw.columns and 'StartHour' in df_raw.columns:
        df_raw['Datetime'] = pd.to_datetime(df_raw['Date'].astype(str) + ' ' + df_raw['StartHour'].astype(str), errors='coerce')
    else:
        # try to find datetime-like
        for col in df_raw.columns:
            try:
                tmp = pd.to_datetime(df_raw[col], errors='coerce')
                if tmp.notna().sum() > 0.6 * len(df_raw):
                    df_raw['Datetime'] = tmp
                    break
            except:
                continue
        if 'Datetime' not in df_raw.columns:
            df_raw['Datetime'] = pd.date_range(start='2000-01-01', periods=len(df_raw), freq='H')
    # drop null production
    if 'Production' in df_raw.columns:
        df_raw = df_raw.dropna(subset=['Production']).reset_index(drop=True)
    else:
        raise KeyError("No 'Production' column found in raw CSV.")
    # pivot by Source
    if 'Source' in df_raw.columns:
        pivot = df_raw.groupby(['Datetime', 'Source'])['Production'].sum().unstack(fill_value=0).reset_index()
    else:
        pivot = df_raw[['Datetime', 'Production']].rename(columns={'Production': 'Total_Production'})
    if 'Wind' not in pivot.columns:
        pivot['Wind'] = 0.0
    if 'Solar' not in pivot.columns:
        pivot['Solar'] = 0.0
    pivot['Total_Production'] = pivot.get('Total_Production', pivot['Wind'] + pivot['Solar'])
    # ensure datetime dtype
    pivot['Datetime'] = pd.to_datetime(pivot['Datetime'])
    pivot = pivot.sort_values('Datetime').reset_index(drop=True)
    pivot['Hour'] = pivot['Datetime'].dt.hour
    pivot['Day'] = pivot['Datetime'].dt.day
    pivot['Month'] = pivot['Datetime'].dt.month
    pivot['DayOfWeek'] = pivot['Datetime'].dt.dayofweek
    df_clean = pivot.copy()
    df_clean.to_csv(CLEANED_PATH, index=False)
    print("Saved cleaned dataset to:", CLEANED_PATH)

print("Data loaded. Shape:", df_clean.shape)
display(df_clean.head())
```

```python
# -------------------------
# 2. Feature Engineering (lags, rolling, cyclic)
# -------------------------
df = df_clean.copy()

# Ensure Datetime index for rolling
df = df.set_index('Datetime').asfreq('H')  # ensure hourly freq; fills missing with NaN
# Forward-fill small gaps and then drop remaining NaNs in Production columns
df[['Wind','Solar','Total_Production']] = df[['Wind','Solar','Total_Production']].fillna(method='ffill').fillna(method='bfill')

# Time features already exist but ensure:
df['Hour'] = df.index.hour
df['Day'] = df.index.day
df['Month'] = df.index.month
df['DayOfWeek'] = df.index.dayofweek

# Lags and rolling means
df['lag_1'] = df['Total_Production'].shift(1)
df['lag_3'] = df['Total_Production'].shift(3)
df['rolling_3h'] = df['Total_Production'].rolling(window=3, min_periods=1).mean().shift(1)
df['rolling_6h'] = df['Total_Production'].rolling(window=6, min_periods=1).mean().shift(1)

# Interaction features
df['solar_hour'] = df['Solar'] * df['Hour']
df['wind_month'] = df['Wind'] * df['Month']

# Cyclic encoding for hour and month
df['hour_sin'] = np.sin(2*np.pi*df['Hour']/24)
df['hour_cos'] = np.cos(2*np.pi*df['Hour']/24)
df['month_sin'] = np.sin(2*np.pi*(df['Month']-1)/12)
df['month_cos'] = np.cos(2*np.pi*(df['Month']-1)/12)

# Drop rows with NaN from lag features
df = df.dropna().reset_index()
print("\nFeature engineering complete. New shape:", df.shape)
display(df.head())
```

```python
# -------------------------
# 3. Prepare dataset for ML models
# -------------------------
feature_cols = ['Hour','Day','Month','DayOfWeek','Wind','Solar','lag_1','lag_3','rolling_3h','rolling_6h',
                'solar_hour','wind_month','hour_sin','hour_cos','month_sin','month_cos']
target_col = 'Total_Production'

X = df[feature_cols].values
y = df[target_col].values

# Train-test split (time-aware: we will split by index to keep sequence for LSTM)
split_idx = int(len(df)*0.8)
X_train_ml = X[:split_idx]
X_test_ml = X[split_idx:]
y_train_ml = y[:split_idx]
y_test_ml = y[split_idx:]

print("ML shapes:", X_train_ml.shape, X_test_ml.shape)

# Scale features for tree models not necessary but for XGBoost often fine; for LSTM we must scale
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_ml)
X_test_scaled = scaler.transform(X_test_ml)

# Save scaler
joblib.dump(scaler, '/content/feature_scaler.pkl')
```

```python
# -------------------------
# 4. Random Forest (baseline)
# -------------------------
rf = RandomForestRegressor(n_estimators=200, random_state=42, n_jobs=-1)
rf.fit(X_train_scaled, y_train_ml)
y_pred_rf = rf.predict(X_test_scaled)
```

```python
# -------------------------
# 5. XGBoost
# -------------------------
xgb = XGBRegressor(n_estimators=200, learning_rate=0.1, random_state=42, verbosity=0, n_jobs=-1)
xgb.fit(X_train_scaled, y_train_ml)
y_pred_xgb = xgb.predict(X_test_scaled)
```

```python
# -------------------------
# 6. LSTM (sequence model)
# -------------------------
# LSTM needs sequences: create windows from original dataframe (not scaled the same way)
from sklearn.preprocessing import MinMaxScaler
seq_scaler = MinMaxScaler()
X_all_for_lstm = df[feature_cols].values
y_all_for_lstm = df[target_col].values.reshape(-1,1)

X_all_scaled = seq_scaler.fit_transform(X_all_for_lstm)
y_all_scaled = seq_scaler.fit_transform(y_all_for_lstm)  # reuse same scaler for target or create separate

# window
WINDOW = 24  # use past 24 hours
def create_sequences(Xs, ys, window=24):
    Xs_seq, ys_seq = [], []
    for i in range(window, len(Xs)):
        Xs_seq.append(Xs[i-window:i])
        ys_seq.append(ys[i])
    return np.array(Xs_seq), np.array(ys_seq)

X_seq, y_seq = create_sequences(X_all_scaled, y_all_scaled, WINDOW)
# split
split_seq = int(len(X_seq)*0.8)
X_train_seq, X_test_seq = X_seq[:split_seq], X_seq[split_seq:]
y_train_seq, y_test_seq = y_seq[:split_seq], y_seq[split_seq:]

print("LSTM shapes:", X_train_seq.shape, X_test_seq.shape)

# Build LSTM model
tf.random.set_seed(42)
model_lstm = Sequential([
    LSTM(64, input_shape=(X_train_seq.shape[1], X_train_seq.shape[2]), return_sequences=False),
    Dropout(0.2),
    Dense(32, activation='relu'),
    Dense(1, activation='linear')
])
model_lstm.compile(optimizer='adam', loss='mse', metrics=['mae'])
es = EarlyStopping(monitor='val_loss', patience=6, restore_best_weights=True)

# Train (keep epochs small to save time; increase if you have time)
history = model_lstm.fit(X_train_seq, y_train_seq, epochs=40, batch_size=32, validation_split=0.1, callbacks=[es], verbose=1)

# Predict and invert scaling
y_pred_seq_scaled = model_lstm.predict(X_test_seq)
# inverse transform: y_all_for_lstm used same scaler -> need to inverse using seq_scaler for target
# Since we scaled y using seq_scaler.fit_transform on y_all_for_lstm using same scaler as features,
# we must inverse transform by constructing a dummy array with same feature count as used earlier.
# Simpler: we scaled y with seq_scaler as well; so we inverse using seq_scaler.inverse_transform
y_pred_lstm = seq_scaler.inverse_transform(y_pred_seq_scaled)
y_test_lstm_true = seq_scaler.inverse_transform(y_test_seq)
```

```python
# -------------------------
# 7. Evaluate models
# -------------------------
def eval_metrics(y_true, y_pred, name="Model"):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"{name} -> MAE: {mae:.2f}, RMSE: {rmse:.2f}, R2: {r2:.4f}")
    return {'mae':mae,'rmse':rmse,'r2':r2}

print("\nEvaluation Results:")
rf_metrics = eval_metrics(y_test_ml, y_pred_rf, "Random Forest")
xgb_metrics = eval_metrics(y_test_ml, y_pred_xgb, "XGBoost")
lstm_metrics = eval_metrics(y_test_lstm_true.flatten(), y_pred_lstm.flatten(), "LSTM")
```

```python
# === Final visualizations + final summary export (Colab-ready) ===
import os, math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import joblib

# helper metrics
def metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = math.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return {'mae':mae, 'rmse':rmse, 'r2':r2}

# Try to load saved CSVs
rf_xgb_path = '/content/test_predictions_rf_xgb.csv'
lstm_path = '/content/test_predictions_lstm.csv'
merged_path = '/content/merged_predictions_all_models.csv'

rf_xgb = pd.read_csv(rf_xgb_path) if os.path.exists(rf_xgb_path) else None
lstm_df = pd.read_csv(lstm_path) if os.path.exists(lstm_path) else None

# If RF/XGB CSV exists but column names vary, try to find Total_Production
def ensure_total(df):
    if df is None: return None
    if 'Total_Production' in df.columns:
        return df
    # try compute from Wind + Solar or from other columns
    if 'Wind' in df.columns and 'Solar' in df.columns:
        df['Total_Production'] = df['Wind'] + df['Solar']
    elif 'Production' in df.columns:
        df['Total_Production'] = df['Production']
    return df

rf_xgb = ensure_total(rf_xgb)
lstm_df = ensure_total(lstm_df)

# If not available, try to reconstruct predictions by loading models & df
recomputed = False
if rf_xgb is None or lstm_df is None:
    print("Some CSVs missing. Attempting to reconstruct from saved models + cleaned CSV (if available)...")
    # load cleaned df if present
    if os.path.exists('/content/cleaned_hybrid_energy.csv'):
        df = pd.read_csv('/content/cleaned_hybrid_energy.csv')
        # ensure feature engineering exists (try-add if missing)
        feature_cols = ['Hour','Day','Month','DayOfWeek','Wind','Solar','lag_1','lag_3','rolling_3h','rolling_6h',
                        'solar_hour','wind_month','hour_sin','hour_cos','month_sin','month_cos']
        if not set(['lag_1','rolling_3h']).issubset(df.columns):
            import numpy as _np
            df['lag_1'] = df['Total_Production'].shift(1).fillna(method='bfill')
            df['lag_3'] = df['Total_Production'].shift(3).fillna(method='bfill')
            df['rolling_3h'] = df['Total_Production'].rolling(3).mean().fillna(method='bfill')
            df['rolling_6h'] = df['Total_Production'].rolling(6).mean().fillna(method='bfill')
            df['solar_hour'] = df['Solar'] * df['Hour']
            df['wind_month'] = df['Wind'] * df['Month']
            df['hour_sin'] = np.sin(2*np.pi*df['Hour']/24)
            df['hour_cos'] = np.cos(2*np.pi*df['Hour']/24)
            df['month_sin'] = np.sin(2*np.pi*(df['Month']-1)/12)
            df['month_cos'] = np.cos(2*np.pi*(df['Month']-1)/12)
        # load models and scaler
        try:
            rf = joblib.load('/content/rf_model.pkl')
            xgb = joblib.load('/content/xgb_model.pkl')
            scaler = joblib.load('/content/standard_scaler.pkl')
            # prepare X and split same way as before
            X = df[feature_cols].values
            y = df['Total_Production'].values
            split_idx = int(len(df)*0.8)
            X_train = X[:split_idx]; X_test = X[split_idx:]
            y_test = y[split_idx:]
            X_test_scaled = scaler.transform(X_test)
            y_pred_rf = rf.predict(X_test_scaled)
            y_pred_xgb = xgb.predict(X_test_scaled)
            # create rf_xgb dataframe
            rf_xgb = df.iloc[split_idx:].copy().reset_index(drop=True)
            rf_xgb['RF_Pred'] = y_pred_rf
            rf_xgb['XGB_Pred'] = y_pred_xgb
            recomputed = True
            print("Recomputed RF/XGB predictions.")
        except Exception as e:
            print("Could not reconstruct RF/XGB:", e)
    else:
        print("cleaned_hybrid_energy.csv not found in /content. Skipping reconstruction.")

# If lstm missing try to reconstruct (needs seq scalers & model)
if lstm_df is None and 'df' in locals():
    try:
        from sklearn.preprocessing import MinMaxScaler
        seq_feat = joblib.load('/content/seq_feat_scaler.pkl')
        seq_targ = joblib.load('/content/seq_target_scaler.pkl')
        from tensorflow.keras.models import load_model
        lstm = load_model('/content/lstm_model.keras') if os.path.exists('/content/lstm_model.keras') else load_model('/content/lstm_model.h5')
        # create sequences and predict on test portion
        WINDOW = 24
        X_all = df[feature_cols].values
        X_all_scaled = seq_feat.transform(X_all)
        # create windows
        def create_seq(Xs, window):
            Xs_seq = []
            for i in range(window, len(Xs)):
                Xs_seq.append(Xs[i-window:i])
            return np.array(Xs_seq)
        X_seq = create_seq(X_all_scaled, WINDOW)
        split_seq = int(len(X_seq)*0.8)
        X_test_seq = X_seq[split_seq:]
        y_pred_scaled = lstm.predict(X_test_seq)
        y_pred_lstm = seq_targ.inverse_transform(y_pred_scaled.reshape(-1,1)).flatten()
        # align with dataframe slices (as earlier logic)
        lstm_test_start_idx = WINDOW + split_seq
        lstm_df = df.iloc[lstm_test_start_idx : lstm_test_start_idx + len(y_pred_lstm)].copy().reset_index(drop=True)
        lstm_df['LSTM_Pred'] = y_pred_lstm
        # true values
        lstm_df['LSTM_Actual'] = df['Total_Production'].iloc[lstm_test_start_idx : lstm_test_start_idx + len(y_pred_lstm)].values
        recomputed = True
        print("Recomputed LSTM predictions.")
    except Exception as e:
        print("Could not reconstruct LSTM:", e)

# Now ensure we have dataframes
if rf_xgb is None:
    raise FileNotFoundError("RF/XGB predictions not available (CSV missing and reconstruction failed).")
if lstm_df is None:
    print("LSTM predictions not available - proceeding with RF/XGB plots only.")

# Merge into a single dataframe for comparison (align by min length)
if lstm_df is not None:
    n = min(len(rf_xgb), len(lstm_df))
    merged = rf_xgb.iloc[:n].reset_index(drop=True).copy()
    merged['LSTM_Pred'] = lstm_df['LSTM_Pred'].iloc[:n].values
    if 'LSTM_Actual' in lstm_df.columns:
        merged['LSTM_Actual'] = lstm_df['LSTM_Actual'].iloc[:n].values
else:
    merged = rf_xgb.copy()
merged.to_csv('/content/merged_predictions_all_models.csv', index=False)
print("Saved merged CSV -> /content/merged_predictions_all_models.csv")

# Compute metrics where possible
summary = {}
if 'RF_Pred' in merged.columns:
    summary['RandomForest'] = metrics(merged['Total_Production'].values, merged['RF_Pred'].values)
if 'XGB_Pred' in merged.columns:
    summary['XGBoost'] = metrics(merged['Total_Production'].values, merged['XGB_Pred'].values)
if 'LSTM_Pred' in merged.columns:
    # LSTM_Actual may be different slice; prefer merged['Total_Production'] alignment
    summary['LSTM'] = metrics(merged['Total_Production'].values, merged['LSTM_Pred'].values)

# Print metrics
print("\nModel Metrics:")
for m, val in summary.items():
    print(f"{m}: MAE={val['mae']:.2f}, RMSE={val['rmse']:.2f}, R2={val['r2']:.4f}")

# ========== VISUALIZATIONS ==========

# 1) Actual vs Predicted (first N samples)
N = min(800, len(merged))
plt.figure(figsize=(14,5))
plt.plot(merged['Total_Production'][:N], label='Actual', linewidth=1)
if 'RF_Pred' in merged.columns: plt.plot(merged['RF_Pred'][:N], label='RF Pred', linewidth=1)
if 'XGB_Pred' in merged.columns: plt.plot(merged['XGB_Pred'][:N], label='XGB Pred', linewidth=1)
if 'LSTM_Pred' in merged.columns: plt.plot(merged['LSTM_Pred'][:N], label='LSTM Pred', linewidth=1)
plt.legend(); plt.title('Actual vs Predicted (sample)'); plt.tight_layout()
plt.savefig('/content/plot_actual_vs_predicted.png', dpi=150)
plt.show()

# 2) Residual histograms
plt.figure(figsize=(12,4))
i=1
if 'RF_Pred' in merged.columns:
    plt.subplot(1,3,i); sns.histplot(merged['Total_Production']-merged['RF_Pred'], bins=50, kde=True); plt.title('RF Residuals'); i+=1
if 'XGB_Pred' in merged.columns:
    plt.subplot(1,3,i); sns.histplot(merged['Total_Production']-merged['XGB_Pred'], bins=50, kde=True); plt.title('XGB Residuals'); i+=1
if 'LSTM_Pred' in merged.columns:
    plt.subplot(1,3,i); sns.histplot(merged['Total_Production']-merged['LSTM_Pred'], bins=50, kde=True); plt.title('LSTM Residuals')
plt.tight_layout()
plt.savefig('/content/plot_residuals.png', dpi=150)
plt.show()

# 3) Scatter Actual vs Predicted with R2 annotation
plt.figure(figsize=(12,4))
if 'RF_Pred' in merged.columns:
    plt.subplot(1,3,1)
    plt.scatter(merged['Total_Production'], merged['RF_Pred'], s=6)
    r = r2_score(merged['Total_Production'], merged['RF_Pred'])
    plt.title(f'RF: Actual vs Pred (R2={r:.3f})'); plt.xlabel('Actual'); plt.ylabel('Pred')
if 'XGB_Pred' in merged.columns:
    plt.subplot(1,3,2)
    plt.scatter(merged['Total_Production'], merged['XGB_Pred'], s=6)
    r = r2_score(merged['Total_Production'], merged['XGB_Pred'])
    plt.title(f'XGB: Actual vs Pred (R2={r:.3f})'); plt.xlabel('Actual'); plt.ylabel('Pred')
if 'LSTM_Pred' in merged.columns:
    plt.subplot(1,3,3)
    plt.scatter(merged['Total_Production'], merged['LSTM_Pred'], s=6)
    r = r2_score(merged['Total_Production'], merged['LSTM_Pred'])
    plt.title(f'LSTM: Actual vs Pred (R2={r:.3f})'); plt.xlabel('Actual'); plt.ylabel('Pred')
plt.tight_layout()
plt.savefig('/content/plot_scatter_actual_vs_pred.png', dpi=150)
plt.show()

# 4) Rolling average comparison (24h)
if len(merged) > 48:
    roll = merged[['Total_Production']].copy()
    roll['RF'] = merged['RF_Pred'] if 'RF_Pred' in merged.columns else np.nan
    roll['XGB'] = merged['XGB_Pred'] if 'XGB_Pred' in merged.columns else np.nan
    roll['LSTM'] = merged['LSTM_Pred'] if 'LSTM_Pred' in merged.columns else np.nan
    roll_sm = roll.rolling(window=24, min_periods=1).mean()
    plt.figure(figsize=(14,5))
    plt.plot(roll_sm['Total_Production'], label='Actual (24h avg)')
    if 'RF_Pred' in merged.columns: plt.plot(roll_sm['RF'], label='RF (24h avg)')
    if 'XGB_Pred' in merged.columns: plt.plot(roll_sm['XGB'], label='XGB (24h avg)')
    if 'LSTM_Pred' in merged.columns: plt.plot(roll_sm['LSTM'], label='LSTM (24h avg)')
    plt.legend(); plt.title('24-hour rolling average: Actual vs Predictions'); plt.tight_layout()
    plt.savefig('/content/plot_rolling_24h.png', dpi=150)
    plt.show()

# 5) Feature importance (Random Forest) if available
if os.path.exists('/content/rf_model.pkl'):
    try:
        rf = joblib.load('/content/rf_model.pkl')
        if hasattr(rf, 'feature_importances_'):
            feat_names = feature_cols
            imp = rf.feature_importances_
            imp_df = pd.DataFrame({'feature':feat_names,'importance':imp}).sort_values('importance',ascending=False)
            plt.figure(figsize=(8,6))
            sns.barplot(x='importance', y='feature', data=imp_df)
            plt.title('RandomForest Feature Importance'); plt.tight_layout()
            plt.savefig('/content/plot_feature_importance.png', dpi=150)
            plt.show()
    except Exception as e:
        print("Could not load RF for feature importance:", e)

# ========== Final summary file ==========
summary_lines = []
summary_lines.append("Hybrid Energy Prediction - Final Summary\n\n")
summary_lines.append("Model metrics:\n")
for k,v in summary.items():
    summary_lines.append(f"{k}: MAE={v['mae']:.2f}, RMSE={v['rmse']:.2f}, R2={v['r2']:.4f}\n")
summary_lines.append("\nSaved files:\n")
for f in ['merged_predictions_all_models.csv','plot_actual_vs_predicted.png','plot_residuals.png','plot_scatter_actual_vs_pred.png','plot_rolling_24h.png','plot_feature_importance.png']:
    if os.path.exists('/content/'+f):
        summary_lines.append(f"- {f}\n")
summary_text = "".join(summary_lines)
with open('/content/final_summary.txt','w') as f:
    f.write(summary_text)

print("\nFinal summary saved to /content/final_summary.txt")
print(summary_text)

# Try to download files (Colab)
try:
    from google.colab import files
    files.download('/content/merged_predictions_all_models.csv')
    files.download('/content/final_summary.txt')
    for f in ['plot_actual_vs_predicted.png','plot_residuals.png','plot_scatter_actual_vs_pred.png','plot_rolling_24h.png','plot_feature_importance.png']:
        if os.path.exists('/content/'+f):
            files.download('/content/'+f)
except Exception as e:
    print("Auto-download failed (not Colab). Files are saved in /content.")
```
