```python
# Full robust preprocessing script for your hybrid energy dataset


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
import warnings
warnings.filterwarnings("ignore")

# ---------- 1. Load dataset (update path if different) ----------
csv_path = '/content/intermittent-renewables-production-france.csv'
df = pd.read_csv(csv_path)
print("Loaded dataset. Columns:", list(df.columns))
print("\nSample rows:")
display(df.head())
```

```python
#Data cleaning
# ---------- 2. Quick missing-values check ----------
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python
# ---------- 3. Create Datetime robustly ----------
def try_create_datetime(df):
    # strategy 1: if 'Datetime' already exists, try to parse it
    if 'Datetime' in df.columns:
        df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
        if df['Datetime'].notna().any():
            return df

    # strategy 2: try 'Date and Hour' column (already looks like "2020-07-22 20:00:00+02:00")
    if 'Date and Hour' in df.columns:
        df['Datetime'] = pd.to_datetime(df['Date and Hour'], errors='coerce')
        if df['Datetime'].notna().any():
            return df

    # strategy 3: combine 'Date' + 'StartHour' into a single string then parse
    if 'Date' in df.columns and 'StartHour' in df.columns:
        def normalize_hour(x):
            if pd.isna(x):
                return ''
            s = str(x).strip()

            if s.isdigit():
                return f"{int(s):02d}:00:00"

            try:
                if '.' in s and s.replace('.', '').isdigit():
                    return f"{int(float(s)):02d}:00:00"
            except:
                pass

            if ':' in s:
                parts = s.split(':')
                if len(parts) == 2:
                    return s + ":00"
                return s
            return s

        combined = df['Date'].astype(str).str.strip() + ' ' + df['StartHour'].apply(normalize_hour)
        df['Datetime'] = pd.to_datetime(combined, errors='coerce')
        if df['Datetime'].notna().any():
            return df

    for col in df.columns:
        try:
            test = pd.to_datetime(df[col], errors='coerce')
            if test.notna().sum() > 0.6 * len(df):  # >60% success
                df['Datetime'] = test
                return df
        except Exception:
            continue

    print("Warning: could not find/parse a datetime column reliably. Creating synthetic datetime index.")
    df['Datetime'] = pd.date_range(start='2000-01-01', periods=len(df), freq='H')
    return df

df = try_create_datetime(df)
print("\nAfter creation: Datetime null count =", int(df['Datetime'].isna().sum()))
```

```python
# ---------- 4. Force Datetime to pandas datetime dtype and remove tz info safely ----------
try:

    df['Datetime'] = pd.to_datetime(df['Datetime'], errors='coerce')
except Exception:

    df['Datetime'] = pd.to_datetime(df['Datetime'].astype(str), utc=True, errors='coerce')


try:
    if df['Datetime'].dt.tz is not None:
        df['Datetime'] = df['Datetime'].dt.tz_convert('UTC').dt.tz_localize(None)
except Exception:

    try:
        df['Datetime'] = pd.to_datetime(df['Datetime'], utc=True, errors='coerce')
        df['Datetime'] = df['Datetime'].dt.tz_convert('UTC').dt.tz_localize(None)
    except Exception:
        if not pd.api.types.is_datetime64_any_dtype(df['Datetime']):
            df['Datetime'] = pd.date_range(start='2000-01-01', periods=len(df), freq='H')

print("Datetime dtype:", df['Datetime'].dtype)
display(df['Datetime'].head(8))
```

```python
# ---------- 5. Drop rows where 'Production' is missing ----------
if 'Production' in df.columns:
    missing_prod = int(df['Production'].isnull().sum())
    print(f"\nRows with missing 'Production': {missing_prod}")
    if missing_prod > 0:
        df = df.dropna(subset=['Production']).reset_index(drop=True)
        print("Dropped rows with missing Production.")
else:
    raise KeyError("Column 'Production' not found. Ensure CSV has 'Production' column.")
```

```python
# ---------- 6. Create time-based features (safe now that Datetime is datetime64) ----------
df['Hour'] = df['Datetime'].dt.hour
df['Day'] = df['Datetime'].dt.day
df['Month'] = df['Datetime'].dt.month
df['DayOfWeek'] = df['Datetime'].dt.dayofweek
```

```python
# ---------- 7. Pivot by Source (create Wind / Solar columns) ----------
if 'Source' in df.columns:

    pivot = df.groupby(['Datetime', 'Source'])['Production'].sum().unstack(fill_value=0).reset_index()
    pivot.columns.name = None
    print("\nPivoted data sample:")
    display(pivot.head(6))
else:
    pivot = df[['Datetime', 'Production']].rename(columns={'Production': 'Total_Production'})
    print("\nNo 'Source' column found; using Production as Total_Production.")
    display(pivot.head())
```

```python
# ---------- 8. Ensure Wind and Solar columns exist, create Total_Production ----------
if 'Wind' in pivot.columns or 'Solar' in pivot.columns:
    if 'Wind' not in pivot.columns:
        pivot['Wind'] = 0.0
    if 'Solar' not in pivot.columns:
        pivot['Solar'] = 0.0
    pivot['Total_Production'] = pivot['Wind'] + pivot['Solar']
else:

    numeric_cols = [c for c in pivot.columns if pd.api.types.is_numeric_dtype(pivot[c]) and c != 'Datetime']
    if 'Total_Production' not in pivot.columns:
        if numeric_cols:
            pivot['Total_Production'] = pivot[numeric_cols].sum(axis=1)
        else:

            pivot = pivot.merge(df[['Datetime','Production']].drop_duplicates('Datetime'), on='Datetime', how='left')
            pivot.rename(columns={'Production':'Total_Production'}, inplace=True)
            pivot['Total_Production'] = pivot['Total_Production'].fillna(0)
```

```python
# ---------- 9. Recreate time features on pivot and sort ----------
pivot = pivot.sort_values('Datetime').reset_index(drop=True)
pivot['Hour'] = pivot['Datetime'].dt.hour
pivot['Day'] = pivot['Datetime'].dt.day
pivot['Month'] = pivot['Datetime'].dt.month
pivot['DayOfWeek'] = pivot['Datetime'].dt.dayofweek

print("\nFinal processed data sample:")
display(pivot.head(6))
```

```python
#feature engineering
# ---------- 10. Simple visualization ----------
plt.figure(figsize=(12,5))
if 'Wind' in pivot.columns and 'Solar' in pivot.columns:
    plt.plot(pivot['Datetime'], pivot['Wind'], label='Wind Production', linewidth=1)
    plt.plot(pivot['Datetime'], pivot['Solar'], label='Solar Production', linewidth=1)
else:
    plt.plot(pivot['Datetime'], pivot['Total_Production'], label='Total Production', linewidth=1)
plt.xlabel('Datetime')
plt.ylabel('Production')
plt.title('Energy Production over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# ---------- 11. Save cleaned dataset ----------
out_path = '/content/cleaned_hybrid_energy.csv'
pivot.to_csv(out_path, index=False)
print(f"\nCleaned data saved to: {out_path}")

# ---------- 12. Quick stats ----------
print("\nBasic stats for Total_Production:")
print(pivot['Total_Production'].describe())

print("\nColumns available for modeling:", list(pivot.columns))
```
